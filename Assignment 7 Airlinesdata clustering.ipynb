```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
data=pd.read_csv(r"C:\Users\Lovely_Ray\Desktop\data science\Assignment 7/EastWestAirlines.csv")
data
```

```python
data.info()
```

```python
data1=data.drop(['ID#'],axis=1)
data1
```

```python
# Normalized data frame (considering the numerical part of data)
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'ward')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
y=pd.DataFrame(hc.fit_predict(data1),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
#adding cluster to dataset
data1['hc_cluster']=Clusters
data1
```

```python
#Inference= We can see the total 3999 observations distributed amongst 0th to 5th clusters with highest number of clusters belonging to cluster no 2
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(data1['hc_cluster'],data1['Balance'], c=hc.labels_) 
```

```python
# Kmeans clustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
airlines=pd.read_csv(r"C:\Users\Lovely_Ray\Desktop\data science\Assignment 7/EastWestAirlines.csv")
airlines
```

```python
airlines.shape
```

```python
airlines.describe()
```

```python
airlines1 =airlines.drop(['ID#'],axis=1)
airlines1
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
from sklearn.preprocessing import StandardScaler

airlines1_norm=StandardScaler().fit_transform(airlines1)
```

```python
airlines1_norm
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(airlines1_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(airlines1_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
airlines2=airlines1.copy()
airlines2['clusters4id']=clusters4.labels_
airlines2
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
airlines2.groupby('clusters4id').agg(['mean']).reset_index()
```

```python
#Plot scatters
plt.figure(figsize=(10, 7))
plt.scatter(airlines2['clusters4id'],airlines2['Balance'], c=clusters4.labels_)
```

```python
clusters6.labels_
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airlines3['clusters6id'],airlines3['Balance'], c=clusters6.labels_)
```
